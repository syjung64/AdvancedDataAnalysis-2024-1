````python
# Pandas 패키기 로드
import pandas as pd
# 설문지 취합 결과 파일 읽어오기
raw_data = pd.read_csv('survey_results_public.csv')
````

````python
# 데이터의 구조를 확인
# 84개 변수, 89,184개 인스턴스
raw_data.info()
````

````python
raw_data.head()
````

````python
# 데이터 분석에 필요한 4개 변수 데이터만 추출
revised_data = raw_data[['Age', 'Country', 'LanguageHaveWorkedWith', 'LearnCode']]
revised_data
````

# 과제 - 1

> 설문에 응답한 개발자들이 어떻게 프로그래밍을 학습했는지 알고싶다

> LearnCode 데이터를 활용하여 백분율로 표기한 파이 그래프를 그려보자

````python
# 설문 응담 개발자들의 프로그래밍 학습 방법 데이터 추출
learn_code = revised_data['LearnCode']
learn_code
                          
````

````python
# 구분자 ';'를 기준으로 학습방법 데이터 분리하여 list로 생성함
learn_code = learn_code.str.split(';')
learn_code.head()
````

````python
# list의 각 데이터를 분리함
# 결과는 series 타입
exploded_learn_code = learn_code.explode()
exploded_learn_code.head(10)
````

````python
exploded_learn_code.unique()
````

````python
size_by_learn_code = exploded_learn_code.to_frame()
size_by_learn_code
````

````python
size_by_learn_code = size_by_learn_code.groupby('LearnCode').size()
size_by_learn_code
````

````python
import matplotlib.pyplot as plt
size_by_learn_code.nlargest(10).plot.pie(figsize = (5, 5), autopct = '%1.2f%%')
plt.ylabel('')
````

# 과제 2
> 과제-1의 파이그래프를 연령대가 25~34세를 대상으로 백분율로 표기한 파이 그래프를 그려보자

````python
learn_for_25_34 = revised_data.query('Age == "25-34 years old"')['LearnCode']
learn_for_25_34.head()
````

````python
learn_for_25_34 = learn_for_25_34.str.split(';').explode()
````

````python
learn_for_25_34.groupby(learn_for_25_34).size().head(5).plot.pie(figsize = (5,5), autopct='%1.2f%%')
````

# Streamlit을 활용한 데이터 분석 애플리케이션

## 소개
Streamlit은 간단하고 빠르게 데이터 분석 애플리케이션을 만들 수 있는 Python 라이브러리입니다. 이를 활용하면 몇 줄의 코드로 인터랙티브한 웹 앱을 구축할 수 있습니다.

## 기능
1. **간단한 코드로 웹 앱 생성**: 몇 줄의 코드로 데이터 시각화 및 대시보드를 구현할 수 있습니다.
2. **다양한 시각화 옵션**: 히스토그램, 선 그래프, 박스 플롯 등 다양한 차트를 손쉽게 통합할 수 있습니다.
3. **사용자 인터페이스 커스터마이징**: 제목, 사이드바, 위젯 등을 쉽게 추가하고 조절할 수 있습니다.

## 예시 코드
```python
import streamlit as st

# 제목
st.title('나의 데이터 분석 앱')

# 데이터 로딩
data = load_data()

# 사이드바에 옵션 추가
options = st.sidebar.multiselect('데이터 시각화 옵션', ['히스토그램', '선 그래프', '박스 플롯'])

# 선택된 옵션에 따라 시각화
if '히스토그램' in options:
    st.bar_chart(data['column_name'])
elif '선 그래프' in options:
    st.line_chart(data)
elif '박스 플롯' in options:
    st.box_plot(data['column_name'])
